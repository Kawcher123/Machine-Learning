```python
import pandas as pd
import numpy as sp
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada.csv")
df
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.year,df.per_capita_incomeUS,color="blue",marker="+")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.per_capita_incomeUS)

reg.predict(2017)
```
