```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('carprice.csv')
df
```

```python
%matplotlib inline
plt.xlabel=""
plt.scatter(df[['Car Model']],df['Sell Price'],marker='*',color="red")
```

```python
dfdummi=pd.get_dummies(df['Car Model'])
dfdummi
```

```python
merged=pd.concat([df,dfdummi],axis='columns')
merged
```

```python
final=merged.drop(['Car Model','Mercedez Benz C class'],axis='columns')
final
```

```python
x=final[['Mileage','Age','Audi A5','BMW X5']]
x
```

```python
y=final[['Sell Price']]
y
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(x,y)
```
